```python
from OCCoursesConfig import pd, plt, sns
%matplotlib inline
import OCCoursesDatasets as ocd
```

```python
# --------------------------------------- GO LIVE
# --- Dataset first load
ocd.load_OC_datasets()
```

```python
def describe_dataset(df,title):
    print("-"*30+" "+title+" : "+str(df.shape[0])+" rows * "+str(df.shape[1])+" columns")
    display(df.head(3))
    print()
#end function

describe_dataset(ocd.OC_Topics,"Topics")

describe_dataset(ocd.OC_Paths,"Paths")
describe_dataset(ocd.OC_PathsSkills,"Paths Skills")

describe_dataset(ocd.OC_Courses,"Courses")
describe_dataset(ocd.OC_CoursesParts,"Courses Parts")
describe_dataset(ocd.OC_CoursesChapters,"Courses Chapters")
describe_dataset(ocd.OC_CoursesSkills,"Courses Skills")
describe_dataset(ocd.OC_CoursesLinks,"Courses Links")

describe_dataset(ocd.OC_Projects,"Projects")
describe_dataset(ocd.OC_ProjectsSkills,"Projects Skills")
describe_dataset(ocd.OC_ProjectsCoursesLinks,"Projects-Courses Links")
```

```python
#cours du topic 3 (data)
df=ocd.OC_Courses[["topic_id","course_id","course_name","course_title","course_language","course_url"]][ocd.OC_Courses["topic_id"].isin([3])]
print("Courses of the 'DATA' path : ",len(df))
display(df.head())
```

```python
# liens inter-cours 
print("links between courses : ",len(ocd.OC_CoursesLinks))
display(ocd.OC_CoursesLinks)
```

```python
# liens projets-cours 
print("links between projects and courses : ",len(ocd.OC_ProjectsCoursesLinks))
display(ocd.OC_ProjectsCoursesLinks)
```

```python
# liens inter-cours qui ont besoin d'un recalcul de cible (href non standard)
df =ocd.OC_CoursesLinks[ocd.OC_CoursesLinks["tgt_course_id"].isin([0])]
unique_hrefs=df["href"].unique()
print("unique values to be resolved : ",len(unique_hrefs))
display(ocd.OC_CoursesLinks[ocd.OC_CoursesLinks["tgt_course_id"].isin([0])].head())
```

```python
# liens inter-cours en doublon (src_course + src_chapter + relation + href)
duplicates = ocd.OC_CoursesLinks[ocd.OC_CoursesLinks.duplicated(['src_course_id','src_chapter_id',"relation","href"], keep=False)]
print("Duplicate links between courses + chapters : ",len(duplicates))
display(duplicates.head())
```

```python
# liens inter-cours en doublon (src_course + href)
duplicates = ocd.OC_CoursesLinks[ocd.OC_CoursesLinks.duplicated(['src_course_id',"href"], keep=False)] 
print("Duplicate links between courses : ",len(duplicates))
display(duplicates.head())
```

```python
# cours qui n'ont pas de lien sortant
courses=set(ocd.OC_Courses["course_id"].values)
links = set(ocd.OC_CoursesLinks["src_course_id"].values)
d=list(courses.difference(links))
print("Courses referencing no other course : ",len(d))
display(ocd.OC_Courses[(ocd.OC_Courses["course_id"].isin(d)) & (ocd.OC_Courses["course_language"].isin(["fr","en","es"]))].head())
```

```python
# cours qui n'ont pas de lien entrant
courses=set(ocd.OC_Courses["course_id"].values)
links = set(ocd.OC_CoursesLinks["tgt_course_id"].values)
d=list(courses.difference(links))
print("Courses referenced by no other course : ",len(d))
display(ocd.OC_Courses[(ocd.OC_Courses["course_id"].isin(d)) & (ocd.OC_Courses["course_language"].isin(["fr","en","es"]))].head())
```

```python
# cours qui n'ont pas de lien, ni entrant ni sortant
courses=set(ocd.OC_Courses["course_id"].values)
out_links = set(ocd.OC_CoursesLinks["src_course_id"].values)
in_links = set(ocd.OC_CoursesLinks["tgt_course_id"].values)
ds=courses.difference(out_links)
d=list(ds.difference(in_links))
print("Courses with no link at all : ",len(d))
display(ocd.OC_Courses[(ocd.OC_Courses["course_id"].isin(d)) & (ocd.OC_Courses["course_language"].isin(["fr","en","es"]))].head())
```
