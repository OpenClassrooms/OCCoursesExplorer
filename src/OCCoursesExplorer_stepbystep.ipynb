## describe datasets

```python
def describe_dataset(df,title):
    print("-"*30+" "+title+" : "+str(df.shape[0])+" rows * "+str(df.shape[1])+" columns")
    display(df.head(3))
    print()
#end function

describe_dataset(ocd.OC_Topics,"Topics")

describe_dataset(ocd.OC_Paths,"Paths")
describe_dataset(ocd.OC_PathsSkills,"Paths Skills")

describe_dataset(ocd.OC_Courses,"Courses")
describe_dataset(ocd.OC_CoursesParts,"Courses Parts")
describe_dataset(ocd.OC_CoursesChapters,"Courses Chapters")
describe_dataset(ocd.OC_CoursesSkills,"Courses Skills")
describe_dataset(ocd.OC_CoursesLinks,"Courses Links")

describe_dataset(ocd.OC_Projects,"Projects")
describe_dataset(ocd.OC_ProjectsSkills,"Projects Skills")
describe_dataset(ocd.OC_ProjectsCoursesLinks,"Projects-Courses Links")

describe_dataset(ocd.OC_MyProgressProjects,"My projects progression")
describe_dataset(ocd.OC_MyProgressCourses,"My courses progression")
describe_dataset(ocd.OC_MyCourses,"My courses exercise dataset")
```

## links between courses

```python
# liens inter-cours 
print("links between courses : ",len(ocd.OC_CoursesLinks))
display(ocd.OC_CoursesLinks)
```

## links between projects and courses

```python
# liens projets-cours 
print("links between projects and courses : ",len(ocd.OC_ProjectsCoursesLinks))
display(ocd.OC_ProjectsCoursesLinks)
```

## find courses with broken links

```python
# liens inter-cours qui ont besoin d'un recalcul de cible (href non standard)
df =ocd.OC_CoursesLinks[ocd.OC_CoursesLinks["tgt_course_id"].isin([0])]
unique_hrefs=df["href"].unique()
print("unique values to be resolved : ",len(unique_hrefs))
display(ocd.OC_CoursesLinks[ocd.OC_CoursesLinks["tgt_course_id"].isin([0])].head())
```

## find duplicate links

```python
# liens inter-cours en doublon (src_course + src_chapter + relation + href)
duplicates = ocd.OC_CoursesLinks[ocd.OC_CoursesLinks.duplicated(['src_course_id','src_chapter_id',"relation","href"], keep=False)]
print("Duplicate links between courses + chapters : ",len(duplicates))
display(duplicates.head())
```

```python
# liens inter-cours en doublon (src_course + href)
duplicates = ocd.OC_CoursesLinks[ocd.OC_CoursesLinks.duplicated(['src_course_id',"href"], keep=False)] 
print("Duplicate links between courses (ignoring linked chapter id): ",len(duplicates))
display(duplicates.head())
```

## courses with no link

```python
# cours qui n'ont pas de lien sortant
courses=set(ocd.OC_Courses["course_id"].values)
links = set(ocd.OC_CoursesLinks["src_course_id"].values)
d=list(courses.difference(links))
print("Courses referencing no other course : ",len(d))
display(ocd.OC_Courses[(ocd.OC_Courses["course_id"].isin(d)) & (ocd.OC_Courses["course_language"].isin(["fr","en","es"]))].head())
```

```python
# cours qui n'ont pas de lien entrant
courses=set(ocd.OC_Courses["course_id"].values)
links = set(ocd.OC_CoursesLinks["tgt_course_id"].values)
d=list(courses.difference(links))
print("Courses referenced by no other course : ",len(d))
display(ocd.OC_Courses[(ocd.OC_Courses["course_id"].isin(d)) & (ocd.OC_Courses["course_language"].isin(["fr","en","es"]))].head())
```

```python
# cours qui n'ont pas de lien, ni entrant ni sortant
courses=set(ocd.OC_Courses["course_id"].values)
out_links = set(ocd.OC_CoursesLinks["src_course_id"].values)
in_links = set(ocd.OC_CoursesLinks["tgt_course_id"].values)
ds=courses.difference(out_links)
d=list(ds.difference(in_links))
print("Courses with no link at all : ",len(d))
display(ocd.OC_Courses[(ocd.OC_Courses["course_id"].isin(d)) & (ocd.OC_Courses["course_language"].isin(["fr","en","es"]))].head())
```

## paths skills

```python
path_id=188

dfp = ocd.OC_Paths[ocd.OC_Paths["path_id"].isin([path_id])][["path_id","path_title"]]
dfs = ocd.OC_PathsSkills
dfps = dfp.merge(dfs,on="path_id",how="left")
print("-"*80)
print(dfps) 
```

## paths without skills

```python
paths=set(ocd.OC_Paths["path_id"].values)
skills = set(ocd.OC_PathsSkills["path_id"].values)
d=list(paths.difference(skills))
print("Paths with no skills : ",len(d))
display(ocd.OC_Paths[(ocd.OC_Paths["path_id"].isin(d))].head())
```

## paths projects skills

```python
path_id=188

dfp = ocd.OC_Paths[ocd.OC_Paths["path_id"].isin([path_id])][["path_id","path_title"]]
dfpp = ocd.OC_Projects[["path_id","project_number","project_title"]]
dfs = ocd.OC_ProjectsSkills[["path_id","project_number","skill"]]
dfps = dfp.merge(dfpp,on="path_id",how="left").merge(dfs,on=["path_id","project_number"],how="left")[["path_title","project_title","skill"]]
print("-"*80)
print(dfps)
 
```

## projects without skills

```python
projects=ocd.OC_Projects.merge(ocd.OC_Paths[["path_id","path_title"]],on="path_id",how="left")[["path_id","project_number","path_title","project_title"]]
skills = ocd.OC_ProjectsSkills[["path_id","project_number"]].drop_duplicates()
delta = projects.merge(skills,on=["path_id","project_number"],how='right', indicator=True)
delta = delta[delta["_merge"]=="left_only"]
print("Projects with no skills : ",len(delta))
display(delta)
#display(delta[delta["_merge"]=="left_only"])
```

## courses skills

```python

dfp = ocd.OC_Paths[ocd.OC_Paths["path_id"].isin([path_id])][["path_id","path_title","topic_id"]]
dfc = ocd.OC_Courses[["course_id","course_title","topic_id"]]
dfs = ocd.OC_CoursesSkills[["course_id","skill"]]
dfcs = dfp.merge(dfc,on="topic_id",how="left").merge(dfs,on="course_id",how="left")[["path_title","course_title","skill"]]
print("-"*80)
print(dfcs)
 
```

## courses without skills

```python
courses=set(ocd.OC_Courses["course_id"].values)
skills = set(ocd.OC_CoursesSkills["course_id"].values)
d=list(courses.difference(skills))
print("Courses with no skills : ",len(d))
display(ocd.OC_Courses[(ocd.OC_Courses["course_id"].isin(d))].head())
display(ocd.OC_Courses[(ocd.OC_Courses["course_id"].isin(d))].course_url.values)
```

## courses parts and chapters

```python

dfp = ocd.OC_Paths[ocd.OC_Paths["path_id"].isin([path_id])][["path_id","path_title","topic_id"]]
dfc = ocd.OC_Courses[["course_id","course_title","topic_id"]]
dfcp = ocd.OC_CoursesParts[["course_id","part_number","part_title"]]
dfcc = ocd.OC_CoursesChapters[["course_id","part_number","chapter_id","chapter_title"]] 
df = dfp.merge(dfc,on="topic_id",how="left").merge(dfcp,on="course_id",how="left").merge(dfcc,on=["course_id","part_number"],how="left")[["path_title","course_title","part_title","chapter_title"]]
 
print("-"*80)
print(df)
 
```
